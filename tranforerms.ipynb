```python
# 导入必要库
from transformers import BertTokenizer, GPT2LMHeadModel

# -------------------------- 配置参数 --------------------------
# 模型名称（必须固定为这个）
model_name = "uer/gpt2-chinese-cluecorpussmall"

# 👉 在这里改你的学号最后一位数字（0-9）
student_id_last_digit = 0

# 根据学号选择对应的开头句子（不要改）
prompt_dict = {
    0: "如果我拥有一台时间机器",
    1: "当人类第一次踏上火星",
    2: "如果动物会说话，它们最想告诉人类的是",
    3: "有一天，城市突然停电了",
    4: "当我醒来，发现自己变成了一本书",
    5: "假如我能隐身一天，我会",
    6: "我走进了那扇从未打开过的门",
    7: "在一个没有网络的世界里",
    8: "如果世界上只剩下我一个人",
    9: "梦中醒来，一切都变了模样"
}

# 获取开头文本
prompt_text = prompt_dict[student_id_last_digit]

# -------------------------- 加载模型 --------------------------
# 关键修复：使用BertTokenizer，与模型匹配
tokenizer = BertTokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# -------------------------- 生成文本 --------------------------
# 编码输入
input_ids = tokenizer.encode(prompt_text, return_tensors="pt")

# 生成设置（防报错、防重复）
output_sequences = model.generate(
    input_ids=input_ids,
    max_length=200,           # 生成长度，可按需调整
    temperature=0.7,          # 随机性，越小越稳定
    top_k=50,                 #  Top-K采样
    repetition_penalty=1.2,     # 重复惩罚
    do_sample=True,           # 开启采样，避免生成死板
    pad_token_id=tokenizer.eos_token_id
)

# 解码输出
generated_text = tokenizer.decode(output_sequences[0], skip_special_tokens=True)

# -------------------------- 打印结果 --------------------------
print("="*50)
print(f"任务要求：基于GPT的文本续写")
print(f"开头句子：{prompt_text}")
print("="*50)
print("\n续写内容：\n")
print(generated_text)
```
